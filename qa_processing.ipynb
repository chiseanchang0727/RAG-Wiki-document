```python
import pandas as pd
import numpy as np
```

```python
qa_pair = pd.read_csv('./data/qa_pair/qa_pair_file.csv')
qa_pair.shape
```

```python
def qa_clean(df):
    df = df.copy()
    df = df.dropna(subset=['Question', 'Answer'])
    df['Answer'] = df['Answer'].str.lower().str.replace(r".", "", regex=False)
    df['ArticleTitle'] = np.where(df['ArticleTitle'].isna(), df['ArticleTitle.1'], df['ArticleTitle'])
    df = df.drop_duplicates(subset=['Question'], keep='first')
    return df
```

```python
qa_pair = qa_clean(qa_pair)
```
